```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Line Plot

```python
x = [1,2,3,4]
y = [12,23,39,45]
plt.title('Line PLot')

plt.plot(x,y)
```

Scatter PLot

```python
plt.scatter(x,y)
# plt.plot(x,y, '+')
```

Bar Chart

```python
company = ['a', 'b', 'c']
revenue = [10, 15, 25]

yposition = np.arange(len(company))
plt.xlabel('companies')
plt.ylabel('revenue')
plt.title('company data')

plt.xticks(yposition, company)
plt.bar(yposition, revenue)
```

Histogram

```python
Marks = [60, 70, 80, 90, 100]
Exp = [10, 12, 14, 18, 20]

plt.xlabel('Marks')
plt.ylabel('EXP')
plt.title('')

plt.hist([Exp, Marks], bins=[60, 80, 100], rwidth=0.9, color=['green', 'orange'])
```
