```python
import pandas as pd
import numpy as np
```

Add a new column to a DataFrame based on existing columns.

```python
data = pd.read_csv('data.csv')
data
```

```python
data['Fit'] = np.where(data['Calories']>400, 'Yes', 'No')
data
```

Rename columns in a DataFrame.

```python
data.rename(columns={"Fit" : "Fitness"}) # Rename columns using a mapping
```

Sort a DataFrame based on values in one or more columns.

```python
sorted_data = data.sort_values(by=['Duration', 'Calories'])
sorted_data.head(20)
```

```python
sorted_data = data.sort_values(by=['Duration', 'Calories'], ascending=False)
sorted_data
```
