```python
import pandas as pd
```

Merge two DataFrames based on a common column.

```python
data1 = pd.DataFrame(
    {
    "A" : [1,2,3,4],
    "B" : [2,3,4,1],
    "C" : [3,4,1,2]
}, index=[0,1,2,3])

data2 = pd.DataFrame(
    {
    "A" : [1,2,3,4],
    "D" : [2,3,4,1],
    "E" : [3,4,1,2]
}, index=[0,5,2,7])

result = pd.merge(data1, data2, on="A")

result
```

Join two DataFrames based on their indices.

```python
data1 = pd.DataFrame(
    {
    "A" : [1,2,3,4],
    "B" : [2,3,4,1],
    "C" : [3,4,1,2]
}, index=[0,1,2,3])

data2 = pd.DataFrame(
    {
    "E" : [1,2,3,4],
    "F" : [2,3,4,1],
    "G" : [3,4,1,2]
}, index=[0,5,2,7])

res = data1.join(data2)
res
```

Concatenate multiple DataFrames along rows and columns.

```python
data1 = pd.DataFrame(
    {
    "A" : [1,2,3,4],
    "B" : [2,3,4,1],
    "C" : [3,4,1,2]
}, index=[0,1,2,3])

data2 = pd.DataFrame(
    {
    "A" : [1,2,3,4],
    "B" : [2,3,4,1],
    "C" : [3,4,1,2]
}, index=[0,5,2,7])

frames = [data1, data2]

result = pd.concat(frames)

result
```
