```python
import pandas as pd
```

Group a DataFrame by a categorical variable and compute summary statistics (e.g., mean, median) for each group.

```python
data = pd.read_csv('data.csv')
```

```python

data.groupby(["Duration"])["Calories"].mean()
```

```python
data.groupby(["Duration"])["Calories"].median()
```

Aggregate data using multiple functions (e.g., mean and count) for each group.

```python
data.Duration.agg(["min", "max", "mean", "count", "sum", "median"])
```

```python
data.groupby(["Duration"]).Calories.agg(["min", "max", "mean", "count", "sum", "median"])
```

Apply custom aggregation functions to grouped data.

```python
custom_agg = {
    'Calories': lambda x: x.max() - x.min(),
    'Duration': lambda x: x.mean()
}

result = data.groupby('Duration').agg(custom_agg)
result
```
